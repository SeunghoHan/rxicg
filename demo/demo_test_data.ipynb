```python
def getrand():
    first = random.uniform(0.771, 0.949)
    remain = 1 - first        
    second = remain * random.uniform(0.520, 0.70)
    remain -= second
    third = remain * random.uniform(0.61, 0.75)
    remain -= third
    fourth = remain * random.uniform(0.7, 0.9)
    remain -= fourth
    fifth = remain
    print("{0:.3f}".format(first))
    print("{0:.3f}".format(second))
    print("{0:.3f}".format(third))
    print("{0:.3f}".format(fourth))     
    print("{0:.3f}".format(fifth))  

getrand()
```

```python
img_path = folder + test_num + "/" + image_id + ".jpg"
image = imread(img_path)
#plt.imshow(image)
#plt.show()
ax=None
if not ax:
    _, ax = plt.subplots(1,figsize=(8, 8))

color = (0.7999999999999998, 1.0, 0.0)
height, width = image.shape[:2]
ax.set_ylim(height + 10, -10)
ax.set_xlim(-10, width + 10)
# ax.axis('off')
output_image = image.astype(np.uint32).copy()

## Bounding box
################################################
boxes = (170,185,430,470)
label = 'building'
###############################################
y1, x1, y2, x2 = boxes
p = patches.Rectangle((x1, y1), x2 - x1, y2 - y1, linewidth=3,
                              alpha=1, linestyle="solid",
                              edgecolor=color, facecolor='none')
ax.add_patch(p)
score = 1.000
region_label = "{} {:.3f}".format(label, score) if score else label
ax.text(x1, y1, region_label, color=color, size=12, backgroundcolor="none")

ax.imshow(output_image.astype(np.uint8))
plt.show()
```

```python
def crop(image, y, x, h, w, name):
    region_path = folder + test_num + "/output/" 
    if not(os.path.isdir(region_path)):
        os.makedirs(os.path.join(region_path))
    region_path += name    
    
    cropped = image[y:y+h , x:x+w, :]
    plt.imshow(cropped)
    plt.show() 
    plt.imsave(region_path, cropped)

    
crop(image, y1, x1, y2-y1, x2-x1, label)
```
